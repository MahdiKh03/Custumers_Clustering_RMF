<h1>Data Cleaning</h1>

```python
cleaned_df = df.copy()
```

```python
cleaned_df["Invoice"] = cleaned_df["Invoice"].astype("str")

mask = (
    cleaned_df["Invoice"].str.match("^\\d{6}$") == True
)
cleaned_df = cleaned_df[mask]
cleaned_df
```

```python
cleaned_df["StockCode"] = cleaned_df["StockCode"].astype("str")
mask = (
    (cleaned_df["StockCode"].str.match("^\\d{5}$") == True)
    | (cleaned_df["StockCode"].str.match("^\\d{5}[a-zA-Z]+$") == True)
    | (cleaned_df["StockCode"].str.match("^PADS$") == True)
)

cleaned_df = cleaned_df[mask]
cleaned_df
```

```python
cleaned_df.dropna(subset=["Customer ID"], inplace=True)
```

```python
cleaned_df = cleaned_df[cleaned_df["Price"] > 0]
```

<h1>Feature Engineering</h1>

```python
cleaned_df["SalesLineTotal"] = cleaned_df["Quantity"] * cleaned_df["Price"]
cleaned_df
```

```python
aggregated_df = cleaned_df.groupby(by="Customer ID", as_index=False).agg(
        MonetaryValues=("SalesLineTotal", "sum"),
        Frequency=("Invoice", "nunique"),
        LastInvoiceDate=("InvoiceDate", "max")
    )
aggregated_df
```

```python
max_invoice_date = aggregated_df['LastInvoiceDate'].max()

aggregated_df["Recency"] = (max_invoice_date - aggregated_df['LastInvoiceDate']).dt.days
aggregated_df.head(10)
```

```python
M_Q1 = aggregated_df["MonetaryValues"].quantile(0.25)
M_Q3 = aggregated_df["MonetaryValues"].quantile(0.75)
M_IQR = M_Q3 - M_Q1

monetary_outliers_df = aggregated_df[(aggregated_df["MonetaryValues"] > (M_Q3 + M_IQR * 1.5)) | (aggregated_df["MonetaryValues"] < ( M_Q1 - M_IQR * 1.5))].copy()
monetary_outliers_df.describe()
```

```python
F_Q1 = aggregated_df["Frequency"].quantile(0.25)
F_Q3 = aggregated_df["Frequency"].quantile(0.75)
F_IQR = F_Q3 - F_Q1

frequency_outliers_df = aggregated_df[(aggregated_df["Frequency"] > (F_Q3 + F_IQR * 1.5)) | (aggregated_df["Frequency"] < ( F_Q1 - F_IQR * 1.5))].copy()
frequency_outliers_df.describe()
```

```python
non_outliers_df = aggregated_df[(~aggregated_df.index.isin(monetary_outliers_df.index)) & (~aggregated_df.index.isin(frequency_outliers_df.index))]
non_outliers_df.describe()
```

```python
plt.figure(figsize=(15,5))

plt.subplot(1, 3, 1)
sns.boxplot(data=non_outliers_df["MonetaryValues"], color="skyblue")
plt.xlabel("Monetary Value")
plt.title('Monetary Value Boxplot')

plt.subplot(1, 3, 2)
sns.boxplot(data=non_outliers_df["Frequency"], color="lightgreen")
plt.xlabel("Frequency")
plt.title('Frequency Boxplot')

plt.subplot(1, 3, 3)
sns.boxplot(data=non_outliers_df["Recency"], color="salmon")
plt.xlabel("Recency")
plt.title('Recency Boxplot')

plt.tight_layout()
plt.show()
```

```python
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(projection="3d")
scatter = ax.scatter(non_outliers_df["MonetaryValues"], non_outliers_df["Frequency"], non_outliers_df["Recency"], color="purple")

ax.set_xlabel("Monetary")
ax.set_ylabel("Frequency")
ax.set_zlabel("Recency")

ax.set_title("3D Scatter Plot of Customer Data")
plt.show()
```

<h1>Standardization</h1>

```python
scaler = StandardScaler()

scaled_data = scaler.fit_transform(non_outliers_df[["MonetaryValues", "Frequency", "Recency"]])
scaled_data
```

```python
scaled_data_df = pd.DataFrame(scaled_data, index=non_outliers_df.index, columns=("MonetaryValues", "Frequency", "Recency"))
scaled_data_df
```

```python
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(projection="3d")
scatter = ax.scatter(scaled_data_df["MonetaryValues"], scaled_data_df["Frequency"], scaled_data_df["Recency"], color="purple")

ax.set_xlabel("Monetary")
ax.set_ylabel("Frequency")
ax.set_zlabel("Recency")

ax.set_title("3D Scatter Plot of Customer Data")
plt.show()
```

<h1>K-Means Clustering</h1>

```python
# Finding the best K 
max_k = 15
inertia = []
silhouette_scores = []
k_values = range(2, max_k + 1)

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42, max_iter=1000)
    cluster_labels = kmeans.fit_predict(scaled_data_df)
    sil_score = silhouette_score(scaled_data_df, cluster_labels)
    silhouette_scores.append(sil_score)
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(14, 6))

plt.subplot(1,2,1)
plt.plot(k_values, inertia, marker="o")
plt.title("KMeans Inertia for Different Values of K")
plt.xlabel('Number of Clusters (k)')
plt.ylabel("Inertia")
plt.xticks(k_values)
plt.grid(True)

plt.subplot(1,2,2)
plt.plot(k_values, silhouette_scores, marker="o", color="red")
plt.title("Silhouette Score for Different Values of K")
plt.xlabel('Number of Clusters')
plt.ylabel("Silhouette Score")
plt.xticks(k_values)
plt.grid(True)

plt.show()
```

```python
# Based on Inertia and Silhouette Score plots the best K is 4
kmeans = KMeans(n_clusters=4, random_state=42, max_iter=1000)
cluster_labels = kmeans.fit_predict(scaled_data_df)
cluster_labels
```

```python
non_outliers_df["Cluster"] = cluster_labels
non_outliers_df
```

<h1>Outliers Analysis</h1>    

```python
overlap_indices = monetary_outliers_df.index.intersection(frequency_outliers_df.index)

monetary_only_outliers = monetary_outliers_df.drop(overlap_indices)
frequency_only_outliers = frequency_outliers_df.drop(overlap_indices)
monetary_and_frequency_outliers = monetary_outliers_df.loc[overlap_indices]

# Manually Assigning Clusters
monetary_only_outliers["Cluster"] = -1
frequency_only_outliers["Cluster"] = -2
monetary_and_frequency_outliers["Cluster"] = -3

outlier_clustering_df = pd.concat([monetary_only_outliers, frequency_only_outliers, monetary_and_frequency_outliers])

outlier_clustering_df
```

```python
cluster_colors = {
    -1: "#eb369c",
    -2: "#3866bc",
    -3: "#126c63"
}

plt.figure(figsize=(12, 18))

plt.subplot(3,1,1)
sns.violinplot(x=outlier_clustering_df["Cluster"], y=outlier_clustering_df["MonetaryValues"], palette=cluster_colors, hue=outlier_clustering_df["Cluster"])
sns.violinplot(y=outlier_clustering_df["MonetaryValues"], color="gray", linewidth=1.0)
plt.title("Monetary Value by Cluster")
plt.ylabel("Monetary Value")
# plt.ylim(None, outlier_clustering_df["MonetaryValues"].max())

plt.subplot(3,1,2)
sns.violinplot(x=outlier_clustering_df["Cluster"], y=outlier_clustering_df["Frequency"], palette=cluster_colors, hue=outlier_clustering_df["Cluster"])
sns.violinplot(y=outlier_clustering_df["Frequency"], color="gray", linewidth=1.0)
plt.title("Frequency by Cluster")
plt.ylabel("Frequency")
# plt.ylim(None, outlier_clustering_df["Frequency"].max())

plt.subplot(3,1,3)
sns.violinplot(x=outlier_clustering_df["Cluster"], y=outlier_clustering_df["Recency"], palette=cluster_colors, hue=outlier_clustering_df["Cluster"])
sns.violinplot(y=outlier_clustering_df["Recency"], color="gray", linewidth=1.0)
plt.title("Recency by Cluster")
plt.ylabel("Recency")
# plt.ylim(None, outlier_clustering_df["Recency"].max())

plt.tight_layout()
plt.show()
```

<h3>Interpreting the Outliers</h3>
<p>Cluster -1 (Monetary Outliers) - Indulge</p>

<p>- Characteristics: High spenders but with infrequent purchases.</p>
<p>- Potential Strategy: Focus on maintaining their loyalty through personalized offers or luxury services tailored to their high spending capacity.</p>
<p>Cluster -2 (Frequency Outliers) - UPSELL</p>

<p>- Characteristics: Frequent buyers who spend less per purchase.</p>
<p>- Potential Strategy: Implement loyalty programs or bundle deals to encourage higher spending per visit.</p>
<p>Cluster -3 (Monetary & Frequency Outliers) - DELIGHT</p>

<p>- Characteristics: Highly valuable customers with both high spending and frequent purchases.</p>
<p>- Potential Strategy: Develop VIP programs or exclusive offers to maintain their loyalty and boost engagement.
This analysis helps identify targeted marketing strategies to better serve each customer group.</p>

```python
cluster_labels = {
    0: "RETAIN",
    1: "RE-ENGAGE",
    2: "NURTURE",
    3: "REWARD",
    -1: "INDULGE",
    -2: "UPSELL",
    -3: "DELIGHT"
}
```

```python
full_clustering_df = pd.concat([non_outliers_df, outlier_clustering_df])

full_clustering_df
```

```python
full_clustering_df["ClusterLabel"] = full_clustering_df["Cluster"].map(cluster_labels)
```
